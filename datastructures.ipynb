The key difference is that **`sort()` changes the list itself** (in-place), while **`sorted()` creates a new list** (leaving the original alone).

Here is the breakdown:

### 1. The Comparison Table

| Feature | `list.sort()` | `sorted(list)` |
| --- | --- | --- |
| **Type** | List **Method** (belongs to the list object) | Built-in **Function** (works on anything) |
| **Effect** | Modifies the **original** list. | Creates a **NEW** list. |
| **Returns** | `None` (It gives back nothing!) | The **new sorted list**. |
| **Works on** | **Lists only.** | Any iterable (Lists, Tuples, Strings, Dictionaries). |

---

### 2. `list.sort()` (The "Organizer")

Think of this like rearranging the books on your shelf. You don't get new books; you just change the order of the ones you have.

* **Best for:** When you don't need the original order anymore and want to save memory.

```python
# Create a messy list
nums = [3, 1, 4, 2]

# Sort it directly
result = nums.sort()

print(nums)   # Output: [1, 2, 3, 4] (The list CHANGED)
print(result) # Output: None (Common mistake! It doesn't return the list)

```

### 3. `sorted()` (The "Photocopier")

Think of this like taking a photo of your bookshelf and rearranging the books *in the photo*. The actual shelf stays messy.

* **Best for:** When you need to keep the original data safe or when sorting things that aren't lists (like tuples).

```python
# Create a messy list
nums = [3, 1, 4, 2]

# Create a NEW sorted version
new_nums = sorted(nums)

print(nums)     # Output: [3, 1, 4, 2] (The original is SAFE)
print(new_nums) # Output: [1, 2, 3, 4] (The new list is sorted)

```

### 4. A Key "Gotcha" (Tuples)

Since `sort()` changes data, it **cannot** work on Tuples (because Tuples are immutable/unchangeable). You **must** use `sorted()` for them.

```python
my_tuple = (3, 1, 2)

# my_tuple.sort()    <-- ERROR! AttributeError
new_list = sorted(my_tuple) # Works! Returns a list: [1, 2, 3]

```


`sort()` works **ONLY** on **Lists**.

It is a specific tool (method) built directly into the List object. If you try to use it on anything else (like a Tuple, String, or Dictionary), Python will give you an error.

### 1. Why only Lists?

The `sort()` method modifies the data **in-place** (it changes the original variable).

* **Tuples and Strings** are **Immutable** (they cannot be changed). Therefore, they cannot support a method that tries to rearrange them.
* **Dictionaries** and **Sets** don't have a specific "order" in the same way lists do (historically), so they don't have this method either.

### 2. The Evidence (Code)

**Works on a List:**

```python
my_list = [3, 1, 2]
my_list.sort()  # ‚úÖ Works perfectly
print(my_list)

```

**Fails on a Tuple:**

```python
my_tuple = (3, 1, 2)
my_tuple.sort()
# ‚ùå AttributeError: 'tuple' object has no attribute 'sort'

```

**Fails on a String:**

```python
my_string = "python"
my_string.sort()
# ‚ùå AttributeError: 'str' object has no attribute 'sort'

```

### Summary Table

| Data Type | Can use `.sort()`? | Why? | What to use instead? |
| --- | --- | --- | --- |
| **List** `[]` | **YES** | It is mutable (changeable). | N/A |
| **Tuple** `()` | **NO** | It is immutable (locked). | `sorted(my_tuple)` |
| **String** `""` | **NO** | It is immutable (locked). | `sorted(my_string)` |
| **Dict** `{}` | **NO** | Sorts by keys usually. | `sorted(my_dict)` |


Here are the code examples for each data type in that table.

### 1. List `[]` (The Only "YES")

Since lists are **mutable** (changeable), `.sort()` works perfectly to rearrange them in place.

```python
# ‚úÖ YES: This works
my_list = [3, 1, 4, 2]

my_list.sort() # Changes the list directly

print(my_list)
# Output: [1, 2, 3, 4]

```

---

### 2. Tuple `()` (NO - Immutable)

Tuples are "locked" after creation. You cannot change their order.

```python
my_tuple = (3, 1, 4, 2)

# ‚ùå NO: This crashes
# my_tuple.sort()
# Error: AttributeError: 'tuple' object has no attribute 'sort'

# ‚úÖ ALTERNATIVE: Use sorted()
new_list = sorted(my_tuple)

print(new_list)
# Output: [1, 2, 3, 4] (Note: It returns a LIST, not a tuple!)

```

---

### 3. String `""` (NO - Immutable)

Strings are also "locked". You can't just swap letters around inside a string.

```python
my_string = "python"

# ‚ùå NO: This crashes
# my_string.sort()
# Error: AttributeError: 'str' object has no attribute 'sort'

# ‚úÖ ALTERNATIVE: Use sorted()
sorted_chars = sorted(my_string)

print(sorted_chars)
# Output: ['h', 'n', 'o', 'p', 't', 'y'] (Returns a LIST of characters)

```

---

### 4. Dictionary `{}` (NO - No Order)

Dictionaries are mapped by keys. `.sort()` doesn't exist for them. When you use `sorted()`, Python sorts the **keys**.

```python
my_dict = {"b": 2, "a": 1, "c": 3}

# ‚ùå NO: This crashes
# my_dict.sort()
# Error: AttributeError: 'dict' object has no attribute 'sort'

# ‚úÖ ALTERNATIVE: Use sorted()
sorted_keys = sorted(my_dict)

print(sorted_keys)
# Output: ['a', 'b', 'c'] (It gives you the sorted KEYS)

```

**Key Takeaway:**
If it's not a List, `my_variable.sort()` will crash. Use `sorted(my_variable)` instead!

#Hashing


Great question üëç
**Hashing** is a **core concept behind sets, dictionaries, fast search, and performance** in Python.
I‚Äôll explain it **slowly, from zero**, with **real-world analogies**, **diagrams (mentally + visually)**, and **Python examples**.

---

## üîπ What is Hashing?

**Hashing** is a technique to **convert data (like a number or string) into a fixed-size number**, called a **hash value**, using a **hash function**.

```text
Input (data) ‚îÄ‚îÄ‚ñ∂ Hash Function ‚îÄ‚îÄ‚ñ∂ Hash Value
```

---

## üß† Real-World Analogy (BEST ONE üî•)

### üìö Library Locker System

* You give your **name** ‚Üí `"Giridhar"`
* System applies a **rule** (hash function)
* Assigns you a **locker number** ‚Üí `42`

üëâ Next time:

* Instead of searching **all lockers**
* It directly jumps to **locker 42**

‚úÖ That‚Äôs **hashing**

---

## üîπ Why Hashing is Needed?

Without hashing (like in a **list**):

* Python checks **one by one**
* Slow for large data

With hashing (like in **set / dict**):

* Python **directly jumps** to the location
* Very fast ‚ö°

---

## üîπ Hashing in Python (Internally)

Python uses hashing in:

* **set**
* **dict**
* **frozenset**
* **keys of dictionary**

Example:

```python
s = {10, 20, 30}
```

When you do:

```python
20 in s
```

Python does **NOT** loop like a list ‚ùå
Instead:

1. Computes `hash(20)`
2. Jumps to the memory location
3. Checks value

‚è± Result: **O(1) average time**

---

## üîπ Visual Idea of Hashing

![Image](https://media.geeksforgeeks.org/wp-content/uploads/20240508162721/Components-of-Hashing.webp)

![Image](https://www.databricks.com/wp-content/uploads/2018/12/hash-buckets.jpg)

![Image](https://sectigostore.com/blog/wp-content/uploads/2020/12/hash-function-in-cryptography-1024x440.png)

Think of memory as **buckets**:

```
Bucket 0 ‚Üí []
Bucket 1 ‚Üí [10]
Bucket 2 ‚Üí []
Bucket 3 ‚Üí [20]
Bucket 4 ‚Üí [30]
```

Hash function decides **which bucket**.

## üîπ Hash Function

A **hash function**:

* Takes input
* Produces an integer
* Same input ‚Üí same hash (always)

```python
print(hash(10))
print(hash("python"))
```

Output will be integers (machine-dependent).

---

## üîπ Important Hash Rules (INTERVIEW üî•)

### Rule 1Ô∏è‚É£ Same value ‚Üí Same hash

```python
hash(10) == hash(10)   # True
```

---

### Rule 2Ô∏è‚É£ Different values ‚Üí May collide

```text
hash("abc") == hash("xyz")   ‚ùì possible
```

This is called **hash collision**.

---

## üîπ What is Hash Collision?

When **two different values produce the same hash**.

Example:

```text
Value A ‚Üí hash = 5
Value B ‚Üí hash = 5
```

Python handles this internally using:

* **Chaining**
* **Open addressing**

üëâ You don‚Äôt manage this manually.

---

## üîπ Why Sets Do NOT Allow Duplicates?

Because of hashing üëá

```python
s = {1, 2, 2, 3}
print(s)   # {1, 2, 3}
```

What happens internally:

1. `hash(2)` already exists
2. Python checks equality
3. Duplicate is ignored

---

## üîπ Why Indexing is NOT Allowed in Sets?

Because:

* Sets are **unordered**
* Data is placed based on **hash**, not position

```python
s = {10, 20, 30}
s[0]   # ‚ùå error
```

---

## üîπ Hashing vs List Search (BIG DIFFERENCE üî•)

### List Search

```python
lst = [1, 2, 3, 4]
4 in lst
```

‚è± Time: **O(n)** (checks one by one)

---

### Set Search

```python
s = {1, 2, 3, 4}
4 in s
```

‚è± Time: **O(1)** (direct access)

---

## üîπ Hashing in Dictionary (dict)

```python
student = {
    "name": "Giridhar",
    "age": 25
}
```

* Keys are **hashed**
* Values are stored at that hash location

```python
print(student["age"])  # instant lookup
```

---

## üîπ Why Only Immutable Types Can Be Hash Keys?

Because hash **must not change**.

‚ùå Mutable ‚Üí hash changes ‚Üí memory breaks

```python
{ [1,2]: "value" }   # ‚ùå error
```

‚úÖ Immutable ‚Üí safe

```python
{ (1,2): "value" }   # ‚úÖ works
```

---
