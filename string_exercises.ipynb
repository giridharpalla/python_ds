```python
from collections import Counter

listw = "Ai is so powerful that ai cant be replace with in the ai"

listw

l = []

l = listw.lower().split(" ")
print(l)

c= Counter(l)

print(dict(c))

#USING FOR LOOP

count = {}
for w in l:
  if w in count:
    count[w]=count[w]+1
  else:
    count[w]=1
print(count)
```

```python
#count the vowels

words = "python is the most powerful langagues"

words = words.replace(" ","")
print(words)

vowels = "aeiou"
count = 0
counter = {}
for w in words:
  if w in vowels:

    count=count+1
    if w in counter:
      counter[w]=counter[w]+1
    else:
      counter[w]=1


print(count)
print(counter)
```

```python
#Reverse a String
text = "Python"
print(text[::-1])   # nohtyP

# Check Palindrome
text = "madam"

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")

# Count Vowels
text = "Python"
count = 0

for char in text.lower():
    if char in "aeiou":
        count += 1

print(count)
```
