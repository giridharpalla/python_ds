# üîπ What is a String in Python?

A **string** is a **sequence of characters** (letters, numbers, symbols) enclosed inside:

* Single quotes `' '`
* Double quotes `" "`
* Triple quotes `''' '''` or `""" """`

### Real-world analogy

üëâ A string is like a **sentence written on paper**.

---

## ‚úÖ Creating Strings

```python
name = "Python"
language = 'Programming'
sentence = """Python is easy to learn"""
```

‚úî All three are valid strings.

---

# üîπ Accessing Characters (Indexing)

Each character has a **position (index)** starting from **0**.

```python
text = "Python"
```

| Index | 0 | 1 | 2 | 3 | 4 | 5 |
| ----- | - | - | - | - | - | - |
| Char  | P | y | t | h | o | n |

```python
print(text[0])   # P
print(text[3])   # h
```

### Negative Indexing

```python
print(text[-1])  # n
print(text[-2])  # o
```

---

# üîπ String Slicing (Substrings)

**Syntax**

```python
string[start : end]
```

> end index is **excluded**

```python
text = "Python Programming"

print(text[0:6])     # Python
print(text[7:18])    # Programming
print(text[:6])      # Python
print(text[7:])      # Programming
```

# üîπ String Immutability (VERY IMPORTANT)

üëâ **Strings cannot be changed directly**

```python
text = "Python"
text[0] = "J"   # ‚ùå ERROR
```

‚úî Instead, create a new string:

```python
text = "Python"
new_text = "J" + text[1:]
print(new_text)  # Jython
```
